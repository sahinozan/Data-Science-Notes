# Matplotlib

```python
import matplotlib.pyplot as plt
```

### Line Plot

```python
year = [1950, 1960, 1970, 1980, 1990]
pop = [1.2, 1.9, 3.1, 5.6, 9.1]
plt.plot(year, pop)  # x axis=year, y axis=pop
plt.show()  # display the plot
```

### Scatter Plot

```python
life_exp = [76, 78, 79, 81, 84]
gdp = [400, 1300, 800, 600, 1400]  # gdp [in billions]
plt.scatter(gdp, life_exp)
plt.show()
```

### Histogram

```python
numbers = [4.1, 3.2, 1.7, 2.6, 3.9, 5.3, 1.8, 2.1, 7.1]
plt.hist(numbers, bins=5)  # bins = how many gap there will be in graph
plt.show()
```

### Decorating the Plot

```python
year = [1950, 1960, 1970, 1980, 1990]
pop = [1.2, 1.9, 3.1, 5.6, 9.1]
plt.plot(year, pop)
plt.xlabel('Year')  # label name of x axis
plt.ylabel('Population')  # label name of y axis
plt.title('Population by Year Projection')  # title of the graph
plt.yticks([0, 2, 4, 6, 8, 10],  # points of y axis
           ['0', '2B', '4B', '6B', '8B', '10B'])  # name of the points on y axis
plt.show()  
```
